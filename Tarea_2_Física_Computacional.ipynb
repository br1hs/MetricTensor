**2. Gráfica de Feigenbaum**



**Responde a las siguientes preguntas**

a) Apoyate en el programa que vimos en clase y escribe un programa que muestre el comportamiento del mapeo logístico mediante una gráfica.

```python
#Mapeo Logístico
#a)
x0=0.5

N=1000

r=4

x=np.zeros(N)
x[0]=x0

for n in range(1,N):
  x[n]=r*x[n-1]*(1-x[n-1])


plt.figure(figsize=(10,4))
plt.plot(range(N), x, 'k', markersize=4, color='darkblue')
plt.title(f"Mapeo logístico (r = {r}, x0 = {x0})")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.show()

```

```python
#Gráica de Feigenbaum
#b)
dr=0.01
r=1

x0=0.5

N=1000

R=[]
X=[]


x=np.zeros(N)
x[0]=x0



while r<4 :
    for n in range(1,N):
      x[n]=r*x[n-1]*(1-x[n-1])
    R.append(r)
    X.append(x[N-1])
    r=r+dr

R_X=np.array([R,X])

plt.figure(figsize=(10,10))
plt.scatter(R_X[0,:],R_X[1,:], color='red')
plt.title("Gráfica de Feigenbaum")
plt.xlabel("r")
plt.ylabel("x[n]")
plt.grid(True)
plt.show()
```

b) De acuerdo con la gráfica, parece que si r es mayor o igual que 3.5 (aproximadamente), la gráfica se vuelve caótica.

```python
#c) "opcional"
N=1000
r=np.array([1.0+0.01*i for i in range(1,N)])
x=np.array([0.5])
for i in range(1,N):
  x=r*x*(1-x)


plt.figure(figsize=(10,10))
plt.scatter( r, x , color='red')
plt.title("Gráfica de Feigenbaum")
plt.xlabel("r")
plt.ylabel("x_n")
plt.grid(True)
plt.show()
```

Ciertamente el programa se optimizó enormemente en compararación con el anterior.

```python
#Conjunto de Mandelbrot


N=1000

X=np.linspace(-2,2,N)
Y=np.linspace(-2,2,N)

Espacio=np.meshgrid(X,Y)

c=Espacio[0]+1j*Espacio[1]

z=np.zeros_like(c,dtype=complex)

Mandelbrot=np.ones(c.shape,dtype=bool)

for i in range (100):
  z=z**2+c
  Mandelbrot[np.abs(z)>2]=False

plt.figure(figsize=(10,10))

plt.imshow(Mandelbrot, extent=(-2,2,-2,2), cmap='gray')
plt.title('Conjunto de Mandelbrot en ')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```
